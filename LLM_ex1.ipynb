```python
print("\n====== CHARACTER LEVEL TOKENIXATION=======")
text= "large language models are transforming artificial intelligence into a new era"
print(text)
char_tokens = list(text)
print(char_tokens[:50],"...")
```

```python
from tokenizers import ByteLevelBPETokenizer

print("\n====== SUB-WORD TOKENIXATION=======")
corpus = [text]

bpe_tokenizer = ByteLevelBPETokenizer()
bpe_tokenizer.train_from_iterator(
    corpus,
    vocab_size=100,
    min_frequency=1
)
bpe_tokens = bpe_tokenizer.encode(text).tokens
print(bpe_tokens)
```

```python
from transformers import BertTokenizer

print("\n===========WORLDPIECE TOKENIXATION===========")
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_tokens = bert_tokenizer.tokenize(text)
print(bert_tokens)
```

```python
!pip install gensim --quiet

from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import nltk

# Download necessary NLTK data
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

# Define corpus
corpus = [
    "large language models are transforming natural language processing",
    "word2vec learns word embeddings from context",
    "natural language understanding is important"
]

print("word2vec")

# Tokenize sentences
sentences = [
    word_tokenize(sentence.lower())
    for sentence in corpus
]

# Train Word2Vec model
w2v_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=1
)

print("Vocabulary:", list(w2v_model.wv.key_to_index.keys()))
print("Vector size:", w2v_model.wv.vector_size)
```
